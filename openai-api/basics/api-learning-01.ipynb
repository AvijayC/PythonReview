```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()

client = OpenAI(api_key=os.getenv("ANTHROPIC_LEARNING_KEY"), base_url="https://api.anthropic.com/v1/")


response = client.completions.create(
    model="claude-opus-4-7",
    prompt="Hello - can you give me an example python script, simple, for saying hello and asking for my name?"
)
```

```python
response = client.chat.completions.create(
    model="claude-opus-4-7",  # Claude model name
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Who are you?"},
    ],
)
```
